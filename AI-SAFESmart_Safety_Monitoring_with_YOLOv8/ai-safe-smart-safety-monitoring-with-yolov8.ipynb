```python
import pandas as pd

data = {
    "Class": ["all", "boots", "gloves", "goggles", "helmet", "person", "vest"],
    "Images": [1669, 442, 279, 604, 534, 306, 572],
    "Labels": [6507, 1276, 653, 875, 1464, 784, 1455],
    "Precision": [0.532, 0.581, 0.459, 0.639, 0.562, 0.377, 0.573],
    "Recall": [0.565, 0.503, 0.352, 0.626, 0.680, 0.579, 0.646],
    "mAP@0.5": [0.509, 0.544, 0.317, 0.655, 0.569, 0.334, 0.635],
    "mAP@0.5:0.95": [0.374, 0.395, 0.218, 0.459, 0.425, 0.243, 0.502]
}

df_metrics = pd.DataFrame(data)

df_metrics
```

# üìä Violation Logging and Statistical Analysis

```python
labels_dir = "/kaggle/working/predict_ppe/labels"

class_names = ['Gloves', 'Goggles', 'Helmet', 'No-Gloves', 'No-Goggles', 'No-Helmet', 'Person']

violation_keywords = ["no-"]
violation_counts = {}

for file in os.listdir(labels_dir):
    if file.endswith(".txt"):
        with open(os.path.join(labels_dir, file), 'r') as f:
            lines = f.readlines()
            for line in lines:
                class_index = int(line.strip().split()[0])
                class_name = class_names[class_index]
                if any(keyword.lower() in class_name.lower() for keyword in violation_keywords):
                    violation_counts[class_name] = violation_counts.get(class_name, 0) + 1

df_violations = pd.DataFrame(list(violation_counts.items()), columns=["Violation Class", "Count"])
df_violations = df_violations.sort_values(by="Count", ascending=False)

if not df_violations.empty:
    plt.figure(figsize=(10,6))
    sns.barplot(data=df_violations, x="Violation Class", y="Count", palette="Reds_r")
    plt.title(" PPE Violation Frequency", fontsize=16)
    plt.xlabel("Violation Type")
    plt.ylabel("Count")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è No violation classes were detected. Please check your tags and class names.")
```

```python
class_names = ['boots', 'gloves', 'goggles', 'helmet', 'person', 'vest']
required_ppe = ['gloves', 'goggles', 'helmet', 'vest'] 

labels_dir = "/kaggle/working/predict_ppe/labels"
image_dir = "/kaggle/input/ppe-dedection/test/images"

violations = []

for label_file in os.listdir(labels_dir):
    if not label_file.endswith(".txt"):
        continue

    img_name = label_file.replace(".txt", ".jpg")

    with open(os.path.join(labels_dir, label_file), "r") as f:
        lines = f.readlines()
        detected_classes = [class_names[int(line.split()[0])] for line in lines]

    for ppe in required_ppe:
        if ppe not in detected_classes:
            violations.append({
                "Image": img_name,
                "Violation Class": f"no-{ppe}"
            })

df_violations = pd.DataFrame(violations)
df_violations.to_csv("/kaggle/working/ppe_violations_log.csv", index=False)
df_violations.head()
```

## üß™ Performance Analysis per Class

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_violations, y="Violation Class", order=df_violations["Violation Class"].value_counts().index, palette="Reds_r")
plt.title("PPE Violation Frequency", fontsize=16)
plt.xlabel("Count")
plt.ylabel("Violation Type")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

## The most violated images (top 10 images)

```python
df_violations["Short Name"] = df_violations["Image"].apply(lambda x: x.split("_jpg")[0] + ".jpg")

top_images = df_violations["Short Name"].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_images.values, y=top_images.index, palette="Oranges_r")
plt.title("Images with Most PPE Violations", fontsize=16)
plt.xlabel("Violation Count")
plt.ylabel("Image")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

# üìà Time-Series and Trend Analysis

```python
import pandas as pd
import numpy as np
import datetime as dt

# üìå Sample date range (7-day simulation)
start_date = dt.datetime(2024, 7, 1, 8)  # Start: July 1st 08:00
timestamps = [start_date + dt.timedelta(minutes=np.random.randint(0, 7*24*60))
              for _ in range(len(df_violations))]

df_violations["Timestamp"] = timestamps
df_violations["Hour"] = df_violations["Timestamp"].dt.hour
df_violations["Date"] = df_violations["Timestamp"].dt.date

df_violations["Timestamp"] = df_violations["Timestamp"].dt.strftime("%Y-%m-%d %H:%M")

df_violations = df_violations[["Short Name", "Violation Class", "Timestamp", "Hour", "Date"]]

styled_df = df_violations.head().style.set_properties(**{
    'font-size': '14px',
    'border-color': 'black',
    'border-width': '1px',
    'text-align': 'center'
}).set_table_styles([
    {'selector': 'th', 'props': [('background-color', '#4E79A7'), ('color', 'white'), ('font-size', '14px')]},
    {'selector': 'td', 'props': [('padding', '6px')]}
]).map(lambda val: 'background-color: #FFB3B3' if isinstance(val, str) and "no-" in val else '')

styled_df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df_violations, x="Hour", palette="coolwarm")
plt.title(" PPE Violations by Hour of Day", fontsize=16)
plt.xlabel("Hour of Day")
plt.ylabel("Number of Violations")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
daily_counts = df_violations.groupby("Date").size()

plt.figure(figsize=(10,5))
daily_counts.plot(marker="o", linestyle="-", color="darkred")
plt.title("Total PPE Violations per Day", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Violations")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import random
from PIL import Image as PILImage

prediction_dir = "/kaggle/working/predict_ppe"
img_size = (224, 224)
grid_rows, grid_cols = 2, 3 
font_size = 8

selected_classes = ["no-helmet", "no-vest", "no-goggles", "no-gloves"]
filtered_df = df_violations[df_violations["Violation Class"].isin(selected_classes)]

matched = []
for idx, row in filtered_df.iterrows():
    short_name = row["Short Name"].replace(".jpg", "")
    for fname in os.listdir(prediction_dir):
        if short_name in fname:
            matched.append({
                "file": fname,
                "violation": row["Violation Class"],
                "timestamp": f"{row['Date']} {row['Hour']}:00"
            })
            break

random.shuffle(matched)
selected = matched[:grid_rows * grid_cols]

fig, axs = plt.subplots(grid_rows, grid_cols, figsize=(grid_cols * 3, grid_rows * 3))
fig.suptitle(" PPE Violations Summary", fontsize=14, color="#D32F2F", weight="bold")

for ax, entry in zip(axs.flat, selected):
    img_path = os.path.join(prediction_dir, entry["file"])
    img = PILImage.open(img_path).resize(img_size)
    ax.imshow(img)
    ax.set_title(f"{entry['timestamp']}", fontsize=font_size, color='darkblue')
    ax.set_xlabel(f" {entry['violation']}", fontsize=font_size, color='darkred')
    ax.axis('off')

for i in range(len(selected), grid_rows * grid_cols):
    axs.flat[i].axis('off')

plt.tight_layout()
plt.show()
```
